```python
import pandas as pd
```

```python
# Read the text file
df = pd.read_csv('SMSSpamCollection.txt', delimiter='\t', names=['label','message'])

# Display the DataFrame
print(df)
```

```python
# Data Cleaning and preprocessing

import re
import nltk
nltk.download('stopwords')
```

### Using PorterStemmer on smsclassifier csv file

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
porterstemmer = PorterStemmer()
```

```python
corpus = []
for i in range (0,len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['message'][i])
    review = review.lower()
    review = review.split()
    review = [porterstemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# For binary BagofWords enable binary=True
cv = CountVectorizer(max_features=2500, binary=True)
```

```python
X = cv.fit_transform(corpus)
```
